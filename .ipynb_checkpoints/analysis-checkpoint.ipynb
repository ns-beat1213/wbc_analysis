# 1. define function

```python
def plot_ind(name):
    sns.set_theme(style="darkgrid")
    sns.set_palette('Dark2')
    # generate individual dataframe
    temp = df[df['Pitcher'] == name]
    
    # groupby dataframe
    #pitch_type = temp['Pitch Type'].value_counts().sort_values(ascending=False)
    g = temp.groupby('Pitch Type')['Pitch Vel  (MPH)'].agg(['count', 'mean', 'std'])
    g2 = temp.groupby('Pitch Type')['Spin (RPM)'].agg(['mean', 'std']).rename(columns = {'mean': 'spin_mean', 'std': 'spin_std'})
    g3 = temp.groupby('Pitch Type')['VBreak'].agg(['mean', 'std']).rename(columns = {'mean': 'vbreak_mean', 'std': 'vbreak_std'})
    g4 = temp.groupby('Pitch Type')['HBreak'].agg(['mean', 'std']).rename(columns = {'mean': 'hbreak_mean', 'std': 'hbreak_std'})

    g_concat = pd.concat([g, g2, g3, g4], axis=1).sort_values('mean', ascending=False)
    
    sum_count = g_concat['count'].sum()
    g_concat['count_per'] = g_concat['count']/sum_count *100
    
    pal = list(g_concat.index)
    
    # plot
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,4.5))
    
    # first, pitch count
    sns.barplot(data=g_concat, y= g_concat.index, x = 'count_per', ax = axes[0], hue_order=pal)
    axes[0].set_title('Pitch Type Distribution')
    axes[0].set_xlabel('Percentage')
    
    # second, scatter plot
    sns.scatterplot(data=g_concat, y= g_concat.index, x = 'mean', ax = axes[1])
    axes[1].errorbar(y = g_concat.index, x = g_concat['mean'], xerr=g_concat['std'], fmt='o')
    axes[1].set_xlim(70,105)
    axes[1].set_xlabel('speed (MPH)')
    axes[1].set_title('Speed (MPH)')
    axes[1].set_yticklabels([])
    
    # Third  Break size plot
    axes[2] = sns.scatterplot(data=temp, x= 'HBreak', y= 'VBreak', hue='Pitch Type',hue_order=pal)
    axes[2].axvline(x=0, color = 'k', linestyle='--', linewidth = 0.7)
    axes[2].axhline(y=-20, color = 'k', linestyle='--', linewidth = 0.7)
    axes[2].set_xlim(-30,30)
    axes[2].set_ylim(-70,10)
    axes[2].set_title(f'Break size')
    
    fig.suptitle(f'{name}', fontsize=16)
    # show
    plt.show()
    return g_concat
```

```python
def show_data(name):
    g = plot_ind(name)
    print('\n')
    print('summary statistics')
    return g.head()
```

```python
name = np.sort(df['Pitcher'].unique())
```

# 2. Interactive plot

```python
@interact_manual(name=name)
def multiple(name=name[0]):
    show_data(name)
```
